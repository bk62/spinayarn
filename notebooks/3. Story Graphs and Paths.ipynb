```python
import spinayarn
from spinayarn.utils import get_fake_llm, draw_graph
from spinayarn import InteractiveStory, SpinAYarn
%matplotlib inline
import networkx as nx

fake_llm = get_fake_llm()
```

### Restaurant

```python
RESTAURANT_OUTLINE = "./examples/restaurant/restaurant.yml"
```

```python
story = InteractiveStory.from_outline_yaml(RESTAURANT_OUTLINE)
spinayarn = SpinAYarn(llm=fake_llm, story_outline=story)
```

```python
spinayarn.prep_generation()
```

```python
spinayarn.story_graph
```

```python
generated = spinayarn.generate_story()
```

```python
from spinayarn.spinayarn import build_story_graph
g = build_story_graph(generated)
```

```python
draw_graph(spinayarn.story_graph)
```

```python
nx.draw_circular(g, with_labels=True, font_size=8)
```
